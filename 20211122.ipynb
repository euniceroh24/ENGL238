```python
amp=1; sr=10000; phase=0

B=493.88
D=587.33
E=659.25
Gb=739.99	
G=783.99	
B5=987.77
```

```python
import numpy as np
import IPython.display as ipd
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
  
```

```python
#from the new world (symphony No.9)

notes=[E, Gb, G, Gb, E, 0, E, 
       E, D, B, D, E, 
       E, Gb, G, Gb, E, 0, E,
       E, G, E, G, B5, B, E ]

durs=[1.0, 0.5, 0.5, 1, 0.5, 0.5, 0.5, 
      0.8, 0.3, 0.2, 0.2, 0.3, 
      1.0, 0.5, 0.5, 1, 0.5, 0.5, 0.5, 
      0.25, 0.2, 0.2, 0.2, 0.6, 0.3, 0.8]

```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
